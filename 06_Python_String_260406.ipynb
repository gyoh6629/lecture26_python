#### 문자열 (str)

```python
s1 = "kopo"
s2 = '한국폴리텍대학 성남캠퍼스'
s3 = '한국폴리텍대학 "성남캠퍼스"'
s4 = '''한국폴리텍대학
성남캠퍼스
인공지능소프트웨어과
        ㅁㄴㅇㄹ'''
print(s1) # kopo
print(s2) # 그대로 나옴
print(s3) # " < 이거 출력 가능, ' 랑 " 구분 해야함
print(s4) # 엔터, 탭 등 다 출력됨, 걍 치는대로 다 가능
```

#### 이스케이프 문자 (escape character)

```python
print('asdf\nasdf') # 엔터, 한줄 내려서 나옴
print('asdf\tasdf') # 탭키 친걸로 나옴
print('asdf\\asdf') # 역슬래시 치려면 \\ 두번 쳐야 하나 나옴
print('asdf\'asdf') # ' 도 역슬래시 치고 쳐야함
print('asdf\"asdf') # " 도 마찬가지
print(r"asdf'\aasdf") # (r"~~") 은 그대로 출력해줌, 줄바꿈, 탭, ', "은 안됨, \만 가능
```

#### 문자열의 연산 : +, *

```python
s = 'Hi! ' + 'Python'
S = 'Hi!' * 3
print(s) # 문자열 끼리 더하기 곱하기 가능, 저장된 변수 이름으로도 가능
print(S)
```

#### 문자열을 원하는 형식으로 만들기 : f-문자열 (Formatted String)

```python
print(f"asdf{1+2}asdf") # { } 안에 변수명, 숫자 다 가능
```

#### { ~~ : }

```python
print(f"{1.123123123:.3f}") # 콜론 뒤에 .3f하면 3자리수까지 나옴, f는 타입임 다른거도 가능
print(f"{1000000000:,}") # 돈 처럼 자리마다 , 생김
print(f"{10000.123123123:,.3f}") # ,.3f 등 둘이 동시 사용 가능
```

## f string

```python
string = 'ABC'
print(f"|{string:10}|")
print(f"|{string:>10}|")
print(f"|{string:^10}|")
print(f"|{string:-^10}|")
print(f"|{string:*>10}|")
```

```python
number = 1234
print(f"|{number:10}|")
print(f"|{number:<10}|")
print(f"|{number:-^10}|")
print(f"|{number:*>10}|")
print(f"|{number:0>10}|")
print(f"|{number:010}|")
```

#### 문자열 다루기 split(), join()

```python
s = 'KAIST,제주대,홍익대'
print(s.split(',')) # split은 기본적으로 한칸씩 띄워줌, ''로 문자 입력하면 그거도 적용해줌, input에도 사용 가능
print(','.join(s)) # 문자열을 하나씩 띄워서 ' ' 안에있는 문자로 이어서 하나의 문자열로 만들어줌, s는 문자열이기 때문에 K,A,I .. 이렇게 나오지만
                   # 리스트로 만들면 인덱스별로 하나씩 연결해줌
print(s.strip('K')) # strip은 문자열 양쪽 끝에서 공백을 제거해줌, 문자 입력하면 그 문자 제거, 중간에 있는건 제거 못함
                    # !! rstrip()은 오른쪽만, lstrip()은 왼쪽만 !!
print(s.replace('K','k')) # replace는 ( a, b ) a 를 b로 바꿔줌
ss = '가나다다라마'
print(ss.find('다')) # find 는 해당 문자의 인덱스를 찾아줌, 보통 그렇듯 왼쪽부터 처음나오는 하나만 > 2
print(ss.rfind('다')) # rfind 는 오른쪽부터 처음 나오는 하나 > 3, 인덱스는 그대로 찾아주는데 왼쪽부터 세냐 오른쪽부터 세냐만 다름
print(ss.count('다')) # 해당 문자열 개수 반환

# str(), int(), float() >> 데이터를 문자열타입, 정수타입, 실수타입으로 바꿔줌

# isdigit() > 문자열이 숫자로만 구성되었는지 확인
# isalpha() > 알파벳으롬나 구성되었는지 확인
# isalnum() > 알파벳과 숫자로만 구성 확인
# islower() > 소문자로만 구성 확인
# isupper() > 대문자
# isspace() > 공백

# lower() > 소문자로 변환
# upper() > 대문자로 변환

# 전부다 .~~()로 사용
```

#### 문자함수 실습

```python
# chr() 아스키코드에 있는대로 정수를 문자로 바꿔줌
# ord() 반대로 정수를 문자로 바꿔줌

# 실습, 알파벳 두 개를 입력 받아서 두 개 사이에 몇 개의 알파벳이 있는지 계산
a, b = input().split()
A, B = ord(a), ord(b)
if A < B : A, B = B, A
print(A-B-1)
```

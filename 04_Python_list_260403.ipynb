### 리스트의 항목에 접근하기
#### 인덱싱(indexing), for문으로 접근하기

```python
test_list = [1, 2, 3, 'one', 'two', 'three']
print(test_list[0]) # 리스트의 인덱스는 0부터 시작 > '1'
print(test_list[3]) # 0 1 2 3 > 'one'
print(test_list[-1]) # 마이너스는 뒤에서 부터 출력함, -1부터 시작 > 'three'
print(test_list[-3]) # -1 -2 -3 > 'one'

for item in test_list: # 리스트 자체도 시퀀스로 사용 가능, 하나씩 순서대로 대입
    print(item, end=" ")
```

### 리스트의 항목에 접근하기
#### 슬라이싱(slicing)

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]
# list [ start : stop : step ] range 랑 같음
print(test_list[:5]) # 0은 안써도 됨, print(test_list[0:5])
print(test_list[1::3]) # start는 비어있으면 0(처음)부터, stop은 비어있으면 끝까지, step은 1씩
print(test_list[:5:2]) # 1 3 5 > 인덱스 0, 2, 4 가 출력된 것이지 1부터 2씩 5까지 출력된게 아님, 인덱스 4값이 5일 뿐
print(test_list[:]) # = [::]
print(test_list[::-1]) # step이 양수면 앞 두개는 비어있을 때 처음과 끝, 음수면 끝과 처음
print(test_list[-1:-7:-2]) # = [::-2]
print(test_list[1:-1])
print(test_list[-3::-1])
```

### 리스트에서 항목 삭제하기
#### pop, remove, del, clear

```python
test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]

item = test_list.pop(2) # pop( 인덱스 ) 2번 인덱스 값 삭제하고 반환(print, return 같은거에 쓸 수 있음)
# print(item) > 3
item = test_list.pop() # 인덱스 안쓰면 젤 뒤에거 삭제하고 반환, (스택에선 top, 맨 마지막(위)에 넣은거)
# print(item) > 맨 뒤 11
test_list.remove('a') # 리스트에서 'a' 값을 찾아서 삭제하는데 여러개 있으면 처음 하나만 삭제, 없으면 에러남
# print(test_list) > 3이랑 11 빠지고 'a'도 삭제돼서 1 2 b c 10

del test_list[2] # 인덱스에 있는 값 삭제
# print(test_list) 추가로 인덱스2번(b) 도 삭제돼서 1 2 c 10
del test_list[:2] # 위에 나온 슬라이싱으로 범위 삭제 가능
# print(test_list) 첨부터 2번 전까지(0 1) 삭제해서 c 10
test_list.clear() # 그냥 다 삭제해서 빈 리스트로 만듦
# print(test_list) 다 삭제해서 []
```

### 리스트 탐색하기
#### in, index, count, len, max, min, sum

```python
test_list = [1, 2, 3, 1, 2, 3]

target = 3
print(target in test_list) # 타겟이 테스트리스트에 있으면 True, 없으면 False 반환, ( 값 in 리스트 )
print(test_list.index(target)) # 타겟 값(3)을 찾아 그 값의 인덱스 번호 반환(처음 하나만 반환함), 없으면 에러남(ValueError)
print(test_list.count(target)) # 리스트 내에 타겟 값(3)이 몇개 있는지 반환

target = 4
print(target in test_list) # 없으므로 False
print(test_list.count(target)) # 없으므로 0
# print(test_list.index(target)) # 리스트에 4가 없으므로 에러

# 프린트 없이 이렇게만 하면 젤 마지막거 하나만 보여줌

len(test_list) # len(리스트) 리스트 크기(값 개수) 반환
max(test_list) # 리스트 안에 값 중 최대값 반환, 타입이 숫자가 아니면( 문자열 등) 에러남
min(test_list) # 최소값, 숫자아니면 에러
sum(test_list) # 값들 모두 다 더함, 위랑 똑같이 에러
```

### 리스트 내포, list comprehension
#### abc = [ i for i in (시퀀스(sequence), range,list 등) if 조건 ]

```python
# 1 ~ N으로 이루어진 리스트 생성
N = int(input())
num_list = []
for i in range(1,N+1):
    num_list.append(i)
print(num_list) # for 로 하나씩 받아서 append함
```

```python
# 1 ~ N으로 이루어진 리스트 생성
N = int(input())
num_list = [i for i in range(1,N+1)] # i 를 추가할건데 i 는 range(1,N+1)로 대입되어 하나씩 추가할거임
print(num_list)

# expression
num_list = [i*2 for i in range(1,N+1)] # 똑같은데 앞에 i에 식을 붙여서 저장할 수 있음, range(1,N+1)에 2씩 곱해서 추가할거임
print(num_list)

# if문 사용
num_list = [i for i in range(1,N+1) if i % 2 == 0] # 똑같은데 i가 2로 나눴을 때 나머지가 0일때만 추가할거임(짝수일때)
print(num_list)
```

### 리스트와 반복구조 연습

### getIndex(num_list, target)
주어진 데이터 target 이 어디에 저장되어 있는지 index를 반환

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getIndex(num_list, target):
    count = 0
    index = 0
    for _ in num_list:
        count += 1
    for i in range(count):
        if target == num_list[i]:
            index = i
    return index

print(getIndex(number_list, 25))
```

### getMax(num_list)
리스트에 저장된 숫자 중 가장 큰 숫자 반환

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getMax(num_list):
    count = 0
    num = num_list[0]
    for _ in num_list:
        count += 1
    for i in range(count):
        if num < num_list[i]:
            num = num_list[i]
    return num
    
print(getMax(number_list))
```

### getMin(num_list)
리스트에 저장된 숫자 중 가장 작은 숫자 반환

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def getMin(num_list):
    count = 0
    num = num_list[0]
    for _ in num_list:
        count += 1
    for i in range(count):
        if num > num_list[i]:
            num = num_list[i]
    return num

print(getMin(number_list))
```

### countGT(num_list, target)
리스트에 저장된 숫자 중 target 보다 큰 수가 몇 개 있는지 구하여 반환

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def countGT(num_list, target):
    count = 0
    countgt = 0
    for _ in num_list:
        count += 1
    for i in range(count):
        if target == num_list[i]:
            countgt += 1
    return countgt

print(countGT(number_list, 42))
```

### sumList(num_list)
리스트에 저장된 모든 값을 더하여 반환

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def sumList(num_list):
    s = 0
    for i in num_list:
        s += i
    return s

print(sumList(number_list))
```

### swapList(num_list)
리스트에 저장된 숫자를 역순으로 저장

```python
number_list = [23, 45, 27, 11, 25, 65, 78]

def swapList(num_list):
    count = 0
    for _ in num_list:
        count += 1
    for i in range(count):
        if i >= count-1-i:
            break
        temp = num_list[i]
        num_list[i] = num_list[count-1-i]
        num_list[count-1-i] = temp
    return num_list
    
swapList(number_list)

print(number_list)
```

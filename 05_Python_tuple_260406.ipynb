#### 튜플 (tuple)의 개념

```python
t1 = (1, 2, 3)                 
t2 = ('a', 'b', ('ab', 'cd'))  # 튜플은 [ , ] 대신 ( , ) 사용
t2 = ('a', 'b', ('ab', 'cd'))  # 리스트랑 전부 같음, 수정만 불가
t3 = ('tuple', (1, 2), [3, 4]) # 리스트 안에 값은 수정 가능, 리스트 칸 자체는 수정 불가
print(t3)
```

#### 튜플(tuple) 만들기

```python
t4 = () # 가능
t5 = (1,) # 값 하나만 넣을 때는 꼭 , 있어야함
t6 = 1, 2, 3 # 튜플은 괄호 없이 , 로만 입력해도 생성 됨
t7 = 1, # 하나 넣을 때 , 는 마찬가지, 튜플은 괄호가 아니라 , 가 정함
print(t7)
x, y, z = t6 # 튜플 값을 여러 변수에 하나씩 저장 가능, 튜플값이 더 많거나 적으면 에러남
print(z)
```

#### 튜플 (tuple) 다루기

```python
# 리스트처럼 쓰기 가능
t8 = 1, 2, 3, 'a', 'b', 'c'
# 튜플 항목에 접근
# print(t8[4]) > b
# print(t8[::2]) > 1, 3, 'b'
# for i in t8:
#    print(i, end="") > 123abc

# 튜플 탐색
# print(3 in t8) > True
# print(t8.index('a')) > 3(인덱스 값임)
# print(t8.count('b')) > 1 (개)

# 튜플 관련 정보
# print(len(t8)) > 6
# print(max(t8[:3])) > 3
# print(min(t8[:3])) > 1
# print(sum(t8[:3])) > 6

# 새로운 튜플 생성
t9 = sorted(t8[3:], reverse = True) # sorted는 튜플 생성이 아니라 무조건 리스트로 나옴, 걍 sorted() 에 튜플 사용할 수 있다는거
# print(t9) > ['c', 'b', 'a']
# print(t8 + tuple(t9)) > 합쳐져서 나옴, t9는 리스트라 튜플로 타입 변경해야 연산 가능
# print(t8*2) > 가능

# 튜플 전체 삭제
# del t9
# print(t9) > t9 삭제돼서 오류남

# 리스트에선 되는데 튜플에선 안되는 것들
# 항목 추가 : append(), insert(), extend()
# 항목 삭제 : pop(), remove(), clear(), del 항목(리스트 안에 항목을 삭제함)
# 항목 재배열 : reverse(), sort()
```

#### 튜플의 활용

```python
x, y = 10, 20
x, y = y, x  # > swap, 이게 튜플 사용했던거

def add_and_mul(a, b):
    return a+b, a*b          # 값 2개 반환
result = add_and_mul(3, 5)
print(result)               # 값 여러개로 반환할 때 튜플로 반환됨
```

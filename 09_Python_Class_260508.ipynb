### 파이썬 클래스

```python
class Circle:
    pi = 3.14
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return self.radius ** 2 * Circle.pi

c = Circle(2.5)
print(c.area())
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    
rec = Rectangle(2.1, 3.4)
print(rec.area())
```

```python
class Account:
    def __init__(self):
        self.balance = 0
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        self.balance -= amount

a1 = Account()
a1.deposit(100)
print(a1.balance)
```

```python
class Account:
    def __init__(self):
        self.__balance = 0
    def deposit(self, amount):
        self.__balance += amount
    def withdraw(self, amount):
        self.__balance -= amount
    def get_balance(self):
        return self.__balance
    
a2 = Account()
a2.deposit(50)
print(a2.get_balance())
```

```python
# 매직 매서드 연습
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f'{self.name}({self.age}살)'
    def __eq__(self, other):
        if self.name == other.name and self.age == other.age:
            return True
        return False

p1 = Person('홍길동', 15)
print(p1)

p2 = Person('홍길동', 15)
print(p2)

if p1 == p2:
    print('같은 사람입니다.')
else:
    print('다른 사람입니다.')
```

```python
# 클래스 Student를 가장 간단히 만든 후, 메소드를 추가하고 객체를 만들어 다음과 같은 결과가
# 나오도록 프로그램을 작성하시오.
class Student:
    def __init__(self, name, univ):
        self.name = name
        self.univ = univ
    def info(self):
        print('대학: ' + self.univ + ' 이름: ' + self.name)
A = Student('이성실', '한국대학교')
A.info()
```

#### 학습 내용 되돌아보기 (4)

```python
# 다음을 만족하는 클래스 Student를 작성해 다음과 같은 결과가 나오도록 프로그램을 작성하시오.
class Student:
    def __init__(self, name, dept, mid, final):
        self.name = name
        self.dept = dept
        self.mid = mid
        self.final = final
    def __str__(self):
        return f'학과: {self.dept}, 이름: {self.name}, 중간: {self.mid}, 기말: {self.final}'
    
    def grade(self):
        avg = int((self.mid + self.final) / 2)
        if avg >= 90:
            return '학점: A'
        elif avg >= 80:
            return '학점: B'
        elif avg >= 70:
            return '학점: C'
        elif avg >= 60:
            return '학점: D'
        else:
            return '학점: F'
A = Student('김경철', '기계학과', 89, 90)
print(A)
print(A.grade())
```

```python
# 클래스 Car에서 다음 조건의 매직 메소드를 만들어 다음과 같은 결과가 나오도록
# 프로그램을 작성하시오.
class Car:
    def __init__(self, company, year, color):
        self.company = company
        self.year = year
        self.color = color
    def __str__(self):
        return f'자동차 회사: {self.company}, 년식: {self.year}, 색상: {self.color}'
    def __eq__(self, other):
        if self.company == other.company and self.year == other.year and self.color == other.color:
            return True
        else:
            return False
mycar = Car('현대', 2020, '검정')
yourcar = Car('기아', 2021, '백색')
print(mycar)
print(yourcar)
print(mycar == yourcar)
```

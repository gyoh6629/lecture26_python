```python
S = input().strip()
result = set()
for i in range(len(S)): # 부분 문자열의 시작 index
    for j in range(i,len(S)): # 부분 문자열의 stop index
        sub_str = S[i:j+1]
        result.add(sub_str)
print(len(result))
```

```python
# import sys
# input = sys.stdin.readline

n = int(input())
in_person = set()

for _ in range(n):
    name, action = input().split()
    if action == 'enter':
        in_person.add(name)
    else:
        in_person.remove(name)

# for person in sorted(in_person, reverse=True):
#     print(person)
print('\n'.join(sorted(in_person, reverse=True)))
```

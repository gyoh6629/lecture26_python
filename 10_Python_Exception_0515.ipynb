```python
try:
    num_list = [2, 6, 34, 65, 11]
    index = int(input('인덱스 입력 : '))
    print(num_list[index])
except (ValueError, IndexError):
    print(f'0부터 {len(num_list)-1}까지의 정수를 입력하세요.')
except Exception as e:
    print(f'예외가 발생했습니다 : {e} -> {type(e)}')
```

```python
try:
    number = int(input('숫자 입력 : '))
except ValueError:
    print('숫자를 입력하지 않았습니다.')
else:
    print(number)
finally:
    print('끝났어요~~~')
```

```python
while True:
    try:
        number = int(input('양의 정수 입력 : '))
        if number <= 0:
            raise ValueError
    except ValueError:
        print('양의 정수를 입력하세요.')
    else:
        sum = 0
        for i in range(1, number+1):
            sum += i
        print(f'>> 1부터 {number}까지의 합 : {sum}')
        break
```

#### 학습 내용 되돌아보기(1)

```python
# 다음 코드에서 출려되는 예외 클래스 이름을 기술하시오.
# (1)
try:
    f = float('30f')
except Exception as e:
    print(type(e)) # ValueError, 문자열을 실수 타입으로 변환 불가, float() 함수에 잘못된 값 입력
# (2)
try:
    {'fb': 11, 'bb': 9, 'vb': 6}['foot']
except Exception as e:
    print(type(e)) # KeyError, 일회용코드고 딕셔너리 만들자 마자 오른쪽 키 조회, 없으므로 키 에러
# (3)
try:
    [1, 2, 3][3]
except Exception as e:
    print(type(e)) # IndexError, 일회용, 인덱스 3 검색, 없으므로 인덱스 에러
# (4)
try:
    pl = 'python' + 3
except Exception as e:
    print(type(e)) # TypeError, 문자열과 정수는 더하기 연산 불가, 더하기 연산에 쓰인 값이 타입 에러
```

#### 학습 내용 되돌아보기 (3)

```python
def divide(x, y):
    try:
        if y == 0:
            raise Exception
        else:
            print('결과:', x/y)
    except:
        print('0으로는 나눌 수 없습니다.')

divide(3.2, 2)
divide(5.4, 0)
```

#### 학습 내용 되돌아보기 (4)

```python
uyear = {1:'freshman', 2:'sophomore', 3:'junior', 4:'senior'}
try:
    grade = int(input('대학교 몇 학년이지요? '))
    if not (1 <= grade <= 4):
        raise Exception
except Exception as e:
    print(f'예외 발생 이름: {type(e)}')
    print('예외 발생 이유: 1~4 정수를 입력하세요.')
else:
    print(f'{grade}학년: {uyear[grade]}')
finally:
    print('예외 처리가 잘되는군요!')
```
